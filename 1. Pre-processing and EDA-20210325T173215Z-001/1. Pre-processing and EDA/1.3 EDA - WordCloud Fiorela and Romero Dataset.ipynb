### WordCloud for Fiorela Dataset

```python
# import csv file
# change file path accordingly 
data = pd.read_csv('/content/gdrive/MyDrive/Depression Project/DepressionDataFinal2Processed.csv')
# read first 5 lines of data
data.head()
```

```python
data.columns
```

```python
# number of null value 
data.isnull().sum()
```

```python
#remove null values
data.dropna(inplace=True)
data.isnull().sum()
```

```python
# filter depressed text 
depressed = data[data['Target']==0]
for index,text in enumerate(depressed["Text"][1:5]):
  print('Tweet %d:\n'%(index+1),text)
```

```python
depressed.shape
```

```python
# filter not depressed text 
notdepressed = data[data['Target']==1]
for index,text in enumerate(notdepressed["Text"][1:5]):
  print('Tweet %d:\n'%(index+1),text)
```

```python
notdepressed.shape
```

```python
# add all the depressed tweets into a list - Fiorela
deptext = ''.join(depressed['Text'].tolist())
```

```python
# # add all the non-depressed tweets into a list - Fiorela 
notdeptext = ''.join(notdepressed['Text'].tolist())
```

```python
#https://www.hiclipart.com/free-transparent-background-png-clipart-dlvwr
# change file path accordingly 
notdepimage = np.array(Image.open('/content/gdrive/MyDrive/Depression Project/thumbs_up.jpg'))
```

```python
#https://www.hiclipart.com/search?clipart=thumbs+down
# change file path accordingly 
depimage = np.array(Image.open('/content/gdrive/MyDrive/Depression Project/thumbs_down.jpg'))
```

```python
# plot Wordcloud for non-depressed tweets - Fiorela
def plot_cloud(wordcloud):
    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud) 
    plt.axis("off");
wordcloud = WordCloud(mask = notdepimage, background_color='#C9F495', random_state=10).generate(notdeptext)
plot_cloud(wordcloud)
```

```python
# plot Wordcloud for depressed tweets - Fiorela
def plot_cloud(wordcloud):
    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud) 
    plt.axis("off");
wordcloud = WordCloud(mask = depimage,  background_color='#F49595', random_state=10).generate(deptext)
plot_cloud(wordcloud)
```

### Evaluating Number of Depressed/ Non-depressed Tweets

```python
# Data to plot
labels = 'Depressive tweets', 'Non-depressive Tweets'
sizes = [4101, 4699]
colors = ['red', 'yellowgreen']
explode = [0.1, 0.0]
```

```python
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Percentage of Depressive and Non-Depressive Tweets in Fiorela Dataset")
plt.axis('equal')
plt.show()
```
